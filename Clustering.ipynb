# Theoretical

Q1. What is unsupervised learning in the context of machine learning?

Unsupervised learning is a type of machine learning where the algorithm learns patterns from unlabeled data.  There are no "correct" answers provided during training. The goal is to discover hidden structures or relationships within the data, such as clusters, associations, or anomalies.

Q2. How does the K-Means clustering algorithm work?

K-Means aims to partition data into K distinct clusters. It works as follows:

1. Initialization: Randomly select K data points as initial cluster centroids.
2. Assignment: Assign each data point to the nearest centroid based on a distance metric (usually Euclidean distance).
3. Update: Recalculate the centroids as the mean of all points assigned to each cluster.
4. Iteration: Repeat steps 2 and 3 until the centroids no longer change significantly or a maximum number of iterations is reached.

Q3. Explain the concept of a dendrogram in hierarchical clustering.

A dendrogram is a tree-like diagram that visually represents the hierarchy of clusters produced by hierarchical clustering. The vertical axis represents the distance or dissimilarity between clusters, and the horizontal axis represents the data points. The height at which two branches merge indicates the distance at which the clusters were merged.

Q4. What is the main difference between K-Means and Hierarchical Clustering?

* K-Means: Partitions data into a pre-specified number (K) of non-overlapping clusters. It's iterative and computationally efficient for large datasets.

* Hierarchical Clustering: Builds a hierarchy of clusters by either iteratively merging (agglomerative) or splitting (divisive) clusters. It produces a dendrogram and doesn't require specifying the number of clusters in advance.

Q5. What are the advantages of DBSCAN over K-Means?

* Handles arbitrary shapes: DBSCAN can find clusters of arbitrary shapes, unlike K-Means which assumes spherical clusters.
* Robust to outliers: DBSCAN can identify and label outliers as noise.
* No need to specify the number of clusters: DBSCAN automatically determines the number of clusters based on data density.

Q6. When would you use Silhouette Score in clustering?

You would use Silhouette Score to evaluate the quality of clusters produced by any clustering algorithm (including K-Means and Hierarchical). It measures how similar each point is to its own cluster compared to other clusters. A higher Silhouette Score indicates better-defined clusters.

Q7. What are the limitations of Hierarchical Clustering?

* Computational cost: Can be computationally expensive for large datasets, especially agglomerative clustering.
* Sensitivity to noise and outliers: Can be sensitive to noise and outliers, leading to poor clustering.
* Difficulty handling high-dimensional data: May not perform well in high-dimensional spaces due to the curse of dimensionality.

Q8. Why is feature scaling important in clustering algorithms like K-Means?

Feature scaling is important because K-Means uses distance-based metrics (e.g., Euclidean distance) to assign data points to clusters. Without scaling, features with larger ranges can dominate the distance calculations, leading to biased clustering.

Q9. How does DBSCAN identify noise points?

DBSCAN identifies noise points as data points that:

* Are not core points (don't have enough neighbors within a specified radius).
* Are not reachable from any core point.

Q10. Define inertia in the context of K-Means.

Inertia in K-Means is the sum of squared distances of samples to their closest cluster center. It represents the within-cluster variance. Lower inertia indicates better clustering (denser clusters).

Q11. What is the elbow method in K-Means clustering?

The elbow method is a heuristic used to determine the optimal number of clusters (K) in K-Means. It involves plotting the inertia (or within-cluster sum of squares) against the number of clusters. The "elbow" point, where the rate of decrease in inertia sharply changes, is considered a good estimate for K.

Q12. Describe the concept of "density" in DBSCAN.

In DBSCAN, density refers to the number of data points within a specified radius (epsilon) around a given point. Core points are points with a high density (at least "minPts" neighbors). Density is used to define clusters and identify noise points.

Q13. Can hierarchical clustering be used on categorical data?

Yes, hierarchical clustering can be used on categorical data. However, you need to use appropriate distance measures for categorical data, such as:

* Jaccard distance: Measures dissimilarity between sets.
* Hamming distance: Measures the number of positions at which corresponding elements are different.
* Gower's distance: Handles mixed data types (numerical and categorical).

Q14. What does a negative Silhouette Score indicate?

A negative Silhouette Score indicates that a data point might be assigned to the wrong cluster. It means the point is closer to points in a neighboring cluster than to points in its own cluster.

Q15. Explain the term "linkage criteria" in hierarchical clustering.

Linkage criteria in hierarchical clustering determine the distance between clusters. Common linkage methods include:

* Single linkage: Minimum distance between points in two clusters.
* Complete linkage: Maximum distance between points in two clusters.
* Average linkage: Average distance between points in two clusters.  
* Ward's method: Minimizes the variance within clusters.

Q16. Why might K-Means clustering perform poorly on data with varying cluster sizes or densities?

K-Means assumes clusters are spherical and have similar densities. It struggles with:

* Varying cluster sizes: Tends to split larger clusters into smaller ones.
* Varying cluster densities: May not accurately identify clusters with different densities.
* Non-spherical shapes: Fails to capture non-linear or complex cluster shapes.

Q17. What are the core parameters in DBSCAN, and how do they influence clustering?

The core parameters in DBSCAN are:

* Epsilon (eps): The radius of the neighborhood around a point. A smaller eps leads to tighter clusters, while a larger eps can merge nearby clusters.
* MinPts: The minimum number of points required to form a dense region (core point). A higher minPts leads to fewer core points and potentially more noise points.

These parameters significantly influence the number and shape of clusters identified by DBSCAN.

Q18. How does K-Means++ improve upon standard K-Means initialization?

K-Means++ improves upon standard K-Means by using a smarter initialization strategy:

1. Select the first centroid randomly.
2. For each remaining data point, calculate the distance to the nearest existing centroid.
3. Choose the next centroid with a probability proportional to the squared distance.
4. Repeat steps 2 and 3 until K centroids are selected.

This helps to spread the initial centroids more evenly, leading to better convergence and reduced chances of getting stuck in local optima.

Q19. What is agglomerative clustering?

Agglomerative clustering is a bottom-up hierarchical clustering method. It starts by treating each data point as a single cluster and iteratively merges the closest pairs of clusters until all points belong to a single cluster or a stopping criterion is met.

Q20. What makes Silhouette Score a better metric than just inertia for model evaluation?

* Inter-cluster separation: Silhouette Score considers both the compactness of clusters (like inertia) and the separation between clusters.
* Handles different shapes: Silhouette Score can be used for clusters of different shapes and sizes, unlike inertia which is more suitable for spherical clusters.
* Provides a range of values: Silhouette Score provides a range of values (-1 to 1) that are easier to interpret than just the inertia value.

Q21. Generate synthetic data with 4 centers using make_blobs and apply K-Means clustering. Visualize using a
scatter plot.

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
```

```python
# Generate synthetic data
X, y = make_blobs(n_samples=300, centers=4, random_state=42)
```

```python
# Apply K-Means clustering
kmeans = KMeans(n_clusters=4, random_state=42)
kmeans.fit(X)
labels = kmeans.labels_
```

```python
# Visualize the clusters
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red', marker='')  # Plot centroids
plt.title('K-Means Clustering')
plt.show()
```

Q29. Create synthetic data using make_blobs and evaluate silhouette scores for k = 2 to 5. Display as a bar chart.

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import numpy as np

# Generate synthetic data
X, y = make_blobs(n_samples=300, centers=4, random_state=42)
```

```python
# Evaluate silhouette scores for different k values
silhouette_scores = []
for k in range(2, 6):
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(X)
    score = silhouette_score(X, labels)
    silhouette_scores.append(score)
```

Q31. Generate synthetic data with overlapping clusters using make_blobs, then apply K-Means and visualize with decision boundaries.

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# Generate synthetic data
X, y = make_blobs(n_samples=300, centers=3, cluster_std=2.0, random_state=42)
```

```python
# Apply K-Means clustering
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)
labels = kmeans.labels_
```

```python
# Visualize the clusters and decision boundaries
h = .02  # Step size of the mesh
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
```

```python
plt.figure(1)
plt.clf()
plt.imshow(Z, interpolation='nearest', extent=(xx.min(), xx.max(), yy.min(), yy.max()), cmap=plt.cm.Paired, aspect='auto', origin='lower')
plt.plot(X[:, 0], X[:, 1], 'k.', markersize=2)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=400, c='red', marker='*')
plt.title('K-Means Clustering with Decision Boundaries')
plt.show()
```

```python
digits = load_digits()
X = digits.data
```

```python
# Apply t-SNE for dimensionality reduction
tsne = TSNE(n_components=2, random_state=42)
X_reduced = tsne.fit_transform(X)
```

```python
# Apply Agglomerative Clustering
agg_clustering = AgglomerativeClustering(n_clusters=10)
labels = agg_clustering.fit_predict(X_reduced)
```

```python
# Visualize the clusters
plt.scatter(X_reduced[:, 0], X_reduced[:, 1], c=labels, cmap='viridis')
plt.title('Agglomerative Clustering (Digits - t-SNE)')
plt.show()
```
